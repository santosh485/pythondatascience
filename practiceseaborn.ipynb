```python
cancer = load_breast_cancer()
```

```python
cancer
```

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
df_cancer
```

```python
sns.scatterplot(x = 'mean texture', y = 'mean radius', hue = 'target', data = df_cancer)
```

```python
sns.countplot(df_cancer['target'], label = "Count")
```

```python
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean area'])
```

```python
plt.figure(figsize = (20, 10))
sns.heatmap(df_cancer.corr(), annot = True)
```

```python
sns.distplot(df_cancer['mean radius'], bins = 24, color = 'blue')
```

```python
sns.scatterplot(x = 'mean radius', y = 'mean area', hue = 'target', data = df_cancer)
```

```python
class_0_df = df_cancer[ df_cancer['target']==0 ]
class_1_df = df_cancer[ df_cancer['target']==1 ]
```

```python
class_0_df
```
