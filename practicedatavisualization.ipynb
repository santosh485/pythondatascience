```python
dail_df = pd.read_csv('daily_returns.csv')
dail_df
```

```python
import random
x = dail_df['AAPL']
y = dail_df['sp500']
plt.scatter(x, y);
```

```python
values  = [20, 55, 5, 17, 3]
colors  = ['g', 'r', 'y', 'b', 'm']
explode = [0, 0.5, 0, 0, 0.2]
labels  = ['AAPL', 'GOOG', 'T', 'TSLA', 'AMZN']
plt.figure(figsize = (7, 7))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('Stock')
plt.show()
```

```python
mu = dail_df['AAPL'].mean()
sigma = dail_df['AAPL'].std()
num_bins = 100

plt.figure(figsize = (5,5))
plt.hist(dail_df['AAPL'], num_bins, facecolor = 'red');
plt.grid()
plt.xlabel('number')
plt.ylabel('Probability')
plt.title("Histogram: mu=" + str(mu) + ', sigma=' + str(sigma))
```

```python
num = 50
plt.hist(dail_df['GOOG'], num, facecolor = 'red');
plt.grid()
```
