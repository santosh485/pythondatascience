```python
age = [5,7,8,7,2,17,2,9,4,11,12,9,6]                  #y=m (slope) x + c (intercept)
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html#scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = stats.linregress(age, speed)

#model predict!
def myfunc(age):
  return slope * age + intercept #speed

mymodel = list(map(myfunc, age)) #M,R,F

''''
This relationship - the coefficient of correlation - is called r.
The r value ranges from 0 to 1, where 0 means no relationship, and 1 means 100% related.
'''
#print("MyModel :",mymodel[::25])
print("Slope: %f  \nIntercept: %f" % (slope, intercept))
print("R-squared: %f" % r_value**2) #high: 
print("R_Value: ",r_value)
print("P_Value: ",p_value)
print("Std_err: ",std_err)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(age, speed)
plt.grid()
plt.plot(age, mymodel)
plt.show()
```

```python
predict_speed = myfunc(15)
print(f"Age=15 > Speeed={predict_speed}")
predict_speed = myfunc(25)
print(f"Age=25 > Speeed={predict_speed}")
predict_speed = myfunc(35)
print(f"Age=35 > Speeed={predict_speed}")
predict_speed = myfunc(5)
print(f"Age=5 > Speeed={predict_speed}")
predict_speed = myfunc(1)
print(f"Age=1 > Speeed={predict_speed}")
```
