```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
X_train, X_test, y_train, y_test = train_test_split(diabetes.data,diabetes.target)
```

```python
X_train= pd.DataFrame(X_train,columns=['age', 'sex', 'bmi', 'map', 'tc', 'ldl', 'hdl', 'tch', 'ltg', 'glu'])
X_test=pd.DataFrame(X_test,columns=['age', 'sex', 'bmi', 'map', 'tc', 'ldl', 'hdl', 'tch', 'ltg', 'glu'])

y_train=pd.DataFrame(y_train,columns=['target'])
y_test=pd.DataFrame(y_test,columns=['target'])
```

```python
age = [5,7,8,7,2,17,2,9,4,11,12,9,6]                  #y=m (slope) x + c (intercept)
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
slope, intercept, r_value, p_value, std_err = stats.linregress(age, speed)
def myfunc(age):
    return slope * age + intercept
mymodel = list(map(myfunc, age))
print("Slope: %f \nIntercept: %f" % (slope, intercept))
print("R-squared: %f" % r_value**2)
print("P_value: ",p_value)
print("Std_err: ",std_err)
```

```python
print(mymodel)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(age, speed)
plt.grid()
plt.plot(age, mymodel)
plt.show()
```

```python
predict_speed = myfunc(15)
print(f"Age=15 > Speed={predict_speed}")
predict_speed = myfunc(25)
print(f"Age=25 > Speed={predict_speed}")
```
